# MATPLOTLIB


````python
import matplotlib.pyplot as plt
````

````python
#plt.plot.scatter("Molecular Weight","measured log solubility in mols per litre")#kind=scatter
import numpy as np
a=np.array(list(df7["Molecular Weight"]))
b=np.array(list(df7["measured log solubility in mols per litre"]))
plt.plot(a,b,'o')
m,c = np.polyfit(a,b,1)
plt.plot(a,m*a+c)
correlation_coefficient = np.corrcoef(a, b)[0, 1]
plt.title(f'Correlation Coefficient: {correlation_coefficient:.2f}')
````

# Results from plot:
-  The negative sign (-) indicates that as one variable increases, the other tends to decrease. In other words, when "Molecular Weight" goes up, "measured log solubility in mols per liter" tends to go down, and vice versa. So, there is a negative trend in the graph.

- The magnitude of correlation coefficient(-0.64) suggests a moderate correlation. -1,1 implises strong correlation. So, here -0.64 implises significant but not so strong relation between the two variabes.

- In practical sense, if we consider a molecule with low molecular weight, the solubility will be comparatively more.
